```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
data['Mental_Health_Condition'].fillna('Unknown', inplace=True)
data['Physical_Activity'].fillna('Unknown', inplace=True)
```

```python
label_encoder = LabelEncoder()
data['Mental_Health_Condition_LabelEncoded'] = label_encoder.fit_transform(data['Mental_Health_Condition'])
data['Physical_Activity_LabelEncoded'] = label_encoder.fit_transform(data['Physical_Activity'])
```

```python
data.isnull().any()
```

```python
onehot_encoder_mental_health = OneHotEncoder(sparse_output=False)
onehot_encoder_physical_activity = OneHotEncoder(sparse_output=False)
```

```python
mental_health_onehot = onehot_encoder_mental_health.fit_transform(data[['Mental_Health_Condition']])
physical_activity_onehot = onehot_encoder_physical_activity.fit_transform(data[['Physical_Activity']])
```

```python
mental_health_categories = onehot_encoder_mental_health.categories_[0]
physical_activity_categories = onehot_encoder_physical_activity.categories_[0]
```

```python
mental_health_df = pd.DataFrame(mental_health_onehot, columns=[f"Mental_Health_{cat}" for cat in mental_health_categories])
physical_activity_df = pd.DataFrame(physical_activity_onehot, columns=[f"Physical_Activity_{cat}" for cat in physical_activity_categories])
```

```python
data_encoded_fixed = pd.concat([data, mental_health_df, physical_activity_df], axis=1)
data_encoded_fixed.head()
```

```python
accuracy = accuracy_score(y_test_class, y_pred_class)
conf_matrix = confusion_matrix(y_test_class, y_pred_class)
```

```python
print(f"Classification - Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix for Stress Level Classification')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
fpr, tpr, _ = roc_curve(y_test_class, y_prob_class)
roc_auc = auc(fpr, tpr)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
precision, recall, _ = precision_recall_curve(y_test_class, y_prob_class)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(recall, precision, color='blue', lw=2)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.grid(True)
plt.show()
```
